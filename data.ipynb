```python
NORMAL_DATA_PATH = "./covid_detection_from_xray/data/NORMAL"
PNEUMONIDA_DATA_PATH = "./covid_detection_from_xray/data/PNEUMONIA"
COVID_DATA_PATH = './covid_detection_from_xray/data/covid-chestxray-dataset'
```

```python
import os
import glob

normal, bacteria, viral, covid = [], [], [], []

n = 0
for f in glob.glob(os.path.join(NORMAL_DATA_PATH, '*')):
  n += 1
  if (n > 5):
      break
  normal.append(f)

n1 = n2 = 0
for f in glob.glob(os.path.join(PNEUMONIDA_DATA_PATH, '*')):
  if ("bacteria" in f):
    n1 += 1
    if (n1 > 5):
      continue
    bacteria.append(f)
  else:
    n2 += 1
    if (n2 > 5):
      continue
    viral.append(f)
  
  if (n1 > 5 and n2 > 5):
    break

n = 0
for f in glob.glob(os.path.join(COVID_DATA_PATH, "*")):
  n += 1
  if (n > 5):
    break
  covid.append(f)
```

```python
import matplotlib.pyplot as plt
from PIL import Image
import cv2
from pylab import rcParams

def plot_image(images):
  rcParams['figure.figsize'] = 50, 50
  fig = plt.figure()

  for i in range(5):
    image = cv2.imread(images[i])

    ax = fig.add_subplot(1, 5, (i + 1))
    ax.imshow(image)

    ax.set_xticks([])
    ax.set_yticks([])
  plt.show()
```

```python
plot_image(normal)
plot_image(bacteria)
plot_image(viral)
plot_image(covid)
```
